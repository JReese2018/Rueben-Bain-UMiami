```python
## Load Libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import euclidean_distances, cosine_similarity
```

```python
## Read Data
df = pd.read_csv('Scout for Defensive Lineman 2023_2024.csv')
df
```

```python
## Cleaning the data
df = df.replace('-', 0)
```

```python
df.corr(numeric_only=True)
```

I was tasked with finding a replacment for Rueben Bain Jr. should he enter the transfer portal.

My first thought is to run a similarity analysis to see what correlates with what. This seems like a good idea to me because features that correlate with each other are usually more relevant.

Based off of my knowledge, this first prediction is going to be off of metrics that I believe are game changing stats 

```python
## Identifying Rueben's Stats
rueben = df[df['Name'] == 'Rueben Bain, Jr.']
```

```python
## Identifying the features I believe are important (excluding percentage stats)
stats = ['Impact Plays', 'Havoc Plays', 'Pressures', 'Sacks', 'Run Stuffs', 'Run Disrupts', 'True Tackles', 'YATA', 'Stops', 'Tot Tackles', 'Time to Pressure']
```

```python
## With Euclidean's Distance
## Identifying Rueben's Stats
rueben = df[df['Name'] == 'Rueben Bain, Jr.']
## Identifying the features I believe are important (excluding percentage stats)
stats = ['Impact Plays', 'Havoc Plays', 'Pressures', 'Sacks', 'Run Stuffs', 'Run Disrupts', 'True Tackles', 'YATA', 'Stops', 'Tot Tackles', 'Time to Pressure']
## Getting the similarity score and adding it the df
similarity_scores = euclidean_distances(rueben[stats], df[stats])
df['Similarity Score'] = similarity_scores[0]
## Showing the top three similar player based off of the stats that I picked
## Sorting the data puts the lowest 
ranked_similarity_df = df.sort_values(by=['Similarity Score'])
ranked_similarity_df = ranked_similarity_df.reset_index()
print(f"Here are the top 5 similar athletes using Euclidean's Distance to Rueben Bain based on the metric given.")
print(f'1. {ranked_similarity_df.Name[1]}')
print(f'2. {ranked_similarity_df.Name[2]}')
print(f'3. {ranked_similarity_df.Name[3]}')
print(f'4. {ranked_similarity_df.Name[4]}')
print(f'5. {ranked_similarity_df.Name[5]}')
```
